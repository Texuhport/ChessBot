```python
#KING'S ASSASSIN
class KingAssassin:
    def __init__(self):
        self.engine_path = "F:\\DemiDev\\Applications\\StockFish\\stockfish\\stockfish-windows-x86-64-avx2.exe"
        self.engine = chess.engine.SimpleEngine.popen_uci(self.engine_path)
        self.depth_check_bonus = {1: 1000, 2: 250, 3: 150, 4: 100, 5: 35}
        self.time_limit = 0.2
        
    def piece_value(self, piece):
        """Assigns value to pieces with a higher penalty for losing non-pawns."""
        if piece is None:
            return 0
        value = {chess.PAWN: 1, chess.KNIGHT: 3, chess.BISHOP: 6, chess.ROOK: 6, chess.QUEEN: 10, chess.KING: 0} 
        return value.get(piece.piece_type, 0)

    def evaluate_move(self, board, move):
        self.start_time = time.time()
        score = 0
        # Make the move on a copy of the board to evaluate its consequences
        board_copy = board.copy(stack=False)
        board_copy.push(move)

        # Reward for putting the opponent's King in check
        if board_copy.is_check():
            score += self.depth_check_bonus[1]

        # Reward for taking higher level pieces and penalize for losing non-pawn pieces
        if board.is_capture(move):
            captured_piece = board_copy.piece_at(move.to_square)
            capturing_piece = board.piece_at(move.from_square)
            score += self.piece_value(captured_piece)

            # Check if the capturing move exposes one of our valuable pieces
            if capturing_piece.piece_type != chess.PAWN:
                score -= self.piece_value(capturing_piece) * 0.699  # Penalize but less than the value of the piece itself
                
        # Evaluate win/loss condition
        if board_copy.is_game_over():
            if board_copy.result() == "1-0" if board.turn == chess.WHITE else "0-1":
                score += 1000  # Greatly reward winning
            else:
                score -= 1000  # Penalize losing or drawing

        
        score += self.assesstheking(board_copy)
        score += self.evaluate_future_checks(board_copy, 1)
        
        return score
        
    def assesstheking(self, board):
        # Count the King's bodyguards
        king_square = board.king(board.turn)
        bodyguards = 0
        safe_position = True  # Assume the king is in a safe position initially

        # Correct calculation for king's rank and file
        king_rank = chess.square_rank(king_square)
        king_file = chess.square_file(king_square)
                 
        for i in range(-2, 3):  # From -3 to 3 including
            for j in range(-2, 3):  # From -3 to 3 including
                if i == 0 and j == 0:
                    continue  # Skip the king's square itself
                neighbor_rank = king_rank + i
                neighbor_file = king_file + j
                
                # Ensure the square is on the board
                if 0 <= neighbor_rank < 8 and 0 <= neighbor_file < 8:
                    neighbor_square = chess.square(neighbor_file, neighbor_rank)
                    piece = board.piece_at(neighbor_square)
                    # Count as a bodyguard if the piece is friendly and not the king
                    if piece and piece.piece_type != chess.KING and piece.color == board.turn:
                        bodyguards += 1
        if board.is_attacked_by(not board.turn, king_square):
            safe_position = False
    
        score_adjustment = 0
        if bodyguards < 5:
            score_adjustment += (5 - bodyguards) * 500
        if not safe_position:
            score_adjustment -= 1000
        return score_adjustment



    def evaluate_future_checks(self, board, current_depth):
        if time.time() - self.start_time > self.time_limit:
            return 0
            
        if current_depth >= 5:  # Base case: stop recursion at depth 5
            return 0
    
        score = 0
        for future_move in board.legal_moves:
            future_board = board.copy(stack=False)
            future_board.push(future_move)
    
            if future_board.is_check():
                score += self.depth_check_bonus[current_depth + 1]
    
            # Recursive call to evaluate further checks at the next depth
            score += self.evaluate_future_checks(future_board, current_depth + 1)
    
        return score

    def evaluate_endgame_move(self, board, move):
        score = 0
        if board.piece_at(move.to_square) and board.piece_at(move.to_square).piece_type == chess.PAWN and (chess.square_rank(move.to_square) in [0, 7]):
            score += 200  # High score for promoting a pawn
        if board.gives_check(move):
            score += 600  # Bonus for moves that give check
        return score
       
    
        # Count the King's bodyguards
        king_square = board.king(board.turn)
        bodyguards = 0
        for i in range(-2, 3):  # From -3 to 3 including
            for j in range(-2, 3):  # From -3 to 3 including
                if i == 0 and j == 0:
                    continue  # Skip the king's square itself
                neighbor_rank = king_rank + i
                neighbor_file = king_file + j
                
                # Ensure the square is on the board
                if 0 <= neighbor_rank < 8 and 0 <= neighbor_file < 8:
                    neighbor_square = chess.square(neighbor_file, neighbor_rank)
                    piece = board.piece_at(neighbor_square)
                    # Count as a bodyguard if the piece is friendly and not the king
                    if piece and piece.piece_type != chess.KING and piece.color == board.turn:
                        bodyguards += 1
        
        # If the King has less than optimal bodyguards, adjust the score to prioritize moving pieces back to protect him
        if bodyguards < 5:
            score += (bodyguards - 2) * 500  # Example adjustment: more points for moves bringing protection
        return score

    def is_endgame(self, board):
        # Initialize piece counts for both colors
        white_pieces_count = 0
        black_pieces_count = 0
    
        # Count the pieces for each color
        for piece_type in chess.PIECE_TYPES:
            white_pieces_count += len(board.pieces(piece_type, chess.WHITE))
            black_pieces_count += len(board.pieces(piece_type, chess.BLACK))
    
        # Calculate the difference in piece count
        piece_count_difference = abs(white_pieces_count - black_pieces_count)
    
        # Define endgame as when one color has an advantage of more than 6 pieces
        return piece_count_difference > 6

    def select_move(self, board, move_history, is_white):
        legal_moves = list(board.legal_moves)
        best_move = None
        best_score = float('-inf')
        endgame = self.is_endgame(board)

        for move in legal_moves:
            if not endgame:
                score = self.evaluate_move(board, move)
                if score > best_score:
                    best_score = score
                    best_move = move
            else:
                score = self.evaluate_endgame_move(board, move)
                if score > best_score:
                    best_score = score
                    best_move = move

        return best_move

    

    def quit_engine(self):
        self.engine.quit()
```

```python
#THE THINKER

class theTHINKER:
    def __init__(self):
        self.engine_path = "F:\\DemiDev\\Applications\\StockFish\\stockfish\\stockfish-windows-x86-64-avx2.exe"
        self.engine = chess.engine.SimpleEngine.popen_uci(self.engine_path)
        self.opening_moves = []
        
    def opponent_ranks(self, is_white):
        return [chess.RANK_1, chess.RANK_2, chess.RANK_3, chess.RANK_4] if not is_white else [chess.RANK_5, chess.RANK_6, chess.RANK_7, chess.RANK_8]

    def get_color(self, is_white):
        # Returns the color 'self' is playing as
        return chess.WHITE if is_white else chess.BLACK

    def count_pieces(self, board):
        # Counts the pieces for both players
        white_pieces = sum(len(board.pieces(piece_type, chess.WHITE)) for piece_type in chess.PIECE_TYPES)
        black_pieces = sum(len(board.pieces(piece_type, chess.BLACK)) for piece_type in chess.PIECE_TYPES)
        white_pawns = len(board.pieces(chess.PAWN, chess.WHITE))
        black_pawns = len(board.pieces(chess.PAWN, chess.BLACK))
        white_non_pawns = sum(len(board.pieces(piece_type, chess.WHITE)) for piece_type in chess.PIECE_TYPES if piece_type != chess.PAWN)
        black_non_pawns = sum(len(board.pieces(piece_type, chess.BLACK)) for piece_type in chess.PIECE_TYPES if piece_type != chess.PAWN)
        return white_pieces, black_pieces, white_pawns, black_pawns, white_non_pawns, black_non_pawns
    
    def phaseCheck(self, is_white):
        white_pieces, black_pieces, white_pawns, black_pawns, white_non_pawns, black_non_pawns = self.count_pieces(board)
        total_pieces = white_pieces + black_pieces
        
        self_pieces = white_pieces if is_white else black_pieces
        opponent_pieces = black_pieces if is_white else white_pieces
        
        if total_pieces > 30:
            phase = "Opening"
        elif total_pieces <= 30:
            if self_pieces > opponent_pieces:
                phase = "MidGame"
            elif self_pieces < opponent_pieces:
                phase = "EndGame"
            else:
                phase = "Limbo"
        
        return phase


    def piece_advantage(self, is_white):
        _, _, white_pawns, black_pawns, white_non_pawns, black_non_pawns = self.count_pieces(board)

        # Determine counts for self and opponent based on the player's color
        self_pawns, self_non_pawns = (white_pawns, white_non_pawns) if is_white else (black_pawns, black_non_pawns)
        opponent_pawns, opponent_non_pawns = (black_pawns, black_non_pawns) if is_white else (white_pawns, white_non_pawns)

        advantage_score = 0

        # Adjust advantage_score based on pawn and non-pawn piece counts
        if self_pawns > opponent_pawns:
            advantage_score += 25
        elif self_pawns < opponent_pawns:
            advantage_score -= 25

        if self_non_pawns > opponent_non_pawns:
            advantage_score += 25
        elif self_non_pawns < opponent_non_pawns:
            advantage_score -= 25

        return advantage_score

    def evaluate_board(self, board, boardphase):
        score = 0
        
        if boardphase == "Opening":
            score += self.knightandbishopopen(board)
        elif boardphase == "MidGame":
            score += self.midgameboardscore(board)
            score += self.RookGate(board)
            score += self.PawnPair(board)
            score += self.PawnSitter(board)
            score += self.activeDefense(board)

 
            
    def evaluate_move(self, board, move, is_white):
        boardphase = self.phaseCheck(is_white)
        board_copy = board.copy(stack=False)
        movescore = 0

        if boardphase == "Opening":
            board_copy.push(move)
            movescore += self.evaluate_board(board_copy, boardphase)
        # Increase score for castling
            if move in board_copy.legal_moves and self.board.is_castling(move):
                movescore += 30
        # Prevent moving the same piece twice
            piece = board_copy.piece_at(move.from_square)
            if piece:
                # Discourage moving the same piece twice
                 if move.from_square in self.opening_moves:
                    movescore -= 50 # Penalize for moving the same piece again    
        #Prevent Queen from leaving the first two ranks in Opening
            queen_square = board_copy.pieces(chess.QUEEN, is_white).pop() if board_copy.pieces(chess.QUEEN, is_white) else None
            
            if queen_square is not None:
                queen_rank = chess.square_rank(queen_square)
            if self.is_white and queen_rank > 1 or not self.is_white and queen_rank < 6:
                movescore -= 25
                
        elif boardphase == "MidGame":
            board_copy.push(move)
            movescore += self.evaluate_board(board_copy, boardphase)
            movescore += self.avengibility(move)
            movescore += self.piece_advantage()
            
        elif boardphase == "EndGame":
            board_copy.push(move)
            movescore += self.evaluate_board(board_copy, boardphase)
            
            # King should not stay in the bottom 2 ranks
            king_square = board_copy.pieces(chess.KING, self.is_white).pop()  # Assuming the king is present
            king_rank = chess.square_rank(king_square)
            if self.is_white and king_rank < 2 or not self.is_white and king_rank > 5:
                movescore -= 20  # Penalize for king's position
            # Queen and Bishops should not stay in the back half ranks
            for piece_type in [chess.QUEEN, chess.BISHOP]:
                for square in board_copy.pieces(piece_type, self.is_white):
                    piece_rank = chess.square_rank(square)
                    if self.is_white and piece_rank > 3 or not self.is_white and piece_rank < 4:
                        movescore -= 10  # Penalize for each Queen/Bishop's position

        return movescore
            
    def select_move(self, board, move_history, is_white):
        best_move = None
        best_score = -1

        for move in board.legal_moves:
            score = self.evaluate_move(board, move, is_white)
            
            # Iteratively check for a better score
            if score > best_score:
                best_score = score
                best_move = move
        self.opening_moves.append(move)
        return best_move
        

    def knightandbishopopen(self, board, is_white):
        score = 0
        for square in chess.SQUARES:
                piece = board.piece_at(square)
                if piece:
                    if piece.color == self.get_color():
                        if piece.piece_type in [chess.KNIGHT, chess.BISHOP]:
                            if self.is_white and chess.square_rank(square) in [5, 6]:
                                score += 10
                            elif not self.is_white and chess.square_rank(square) in [2, 3]:
                                score += 10
        return score

    def midgameboardscore(self, board):
        square_values = {
            chess.D4: 10, chess.D5: 10, chess.E4: 10, chess.E5: 10,  # Most central squares
            chess.C4: 8, chess.C5: 8, chess.F4: 8, chess.F5: 8,      # Immediate left and right
            chess.B4: 6, chess.B5: 6, chess.G4: 6, chess.G5: 6,      # Two squares away
            chess.A4: 4, chess.A5: 4, chess.H4: 4, chess.H5: 4,      # Edge squares
        }
        
        score = 0
        for square, value in square_values.items():
            piece = board.piece_at(square)
            if piece:
                if piece.color == self.get_color():
                    score += value
                else:
                    score -= value 

        return score
        
    def RookGate(self, board):
        rook_squares = list(board.pieces(chess.ROOK, self.get_color()))
        if len(rook_squares) < 2:
            # If there are less than 2 rooks, no points are awarded
            return 0

        # Extract ranks and files of both rooks
        rook_positions = [(chess.square_rank(square), chess.square_file(square)) for square in rook_squares]

        # Check if the rooks are on the same rank or file
        rook1, rook2 = rook_positions
        if rook1[0] == rook2[0] or rook1[1] == rook2[1]:
            return 25
        else:
            return 0
        
    def PawnPair(self, board):
        pawn_pairs = 0
        for rank in range(8):  
            previous_square_had_pawn = False
            for file in range(8):  
                square = chess.square(file, rank)
                piece = self.board.piece_at(square)
                if piece and piece.piece_type == chess.PAWN and piece.color == self.get_color():
                    if previous_square_had_pawn:
                        pawn_pairs += 1
                        previous_square_had_pawn = False  # Reset after counting a pair
                    else:
                        previous_square_had_pawn = True
                else:
                    previous_square_had_pawn = False  # Reset if no pawn or pawn of opposite color

        # Apply penalty for each pawn pair
        penalty_per_pair = -15  # Define the penalty for each pair
        score_penalty = pawn_pairs * penalty_per_pair
        return score_penalty

    def PawnSitter(self, board):
        score_bonus = 0
        opponent_ranks = self.opponent_ranks()
        for file in range(8):  # Loop through all files
            pawn_found = False
            rook_found = False
            for rank in opponent_ranks:  # Only check opponent's territory
                square = chess.square(file, rank)
                piece = self.board.piece_at(square)
                if piece and piece.color == self.get_color(is_white):
                    if piece.piece_type == chess.PAWN:
                        pawn_found = True
                    elif piece.piece_type == chess.ROOK:
                        rook_found = True
            if pawn_found and rook_found:
                score_bonus += 20
        return score_bonus

    def activeDefense(self, board):
        score = 0
        # Identify all pieces that are important (King, Queen, Rooks)
        important_pieces_squares = board.pieces(chess.KING, self.get_color(is_white)) | \
                                   board.pieces(chess.QUEEN, self.get_color(is_white)) | \
                                   board.pieces(chess.ROOK, self.get_color(is_white))

        # Check each piece of the player
        for piece_type in chess.PIECE_TYPES:
            for square in board.pieces(piece_type, self.get_color()):
                # Generate all possible opponent moves
                for move in board.legal_moves:
                    if move.to_square in important_pieces_squares and board.is_capture(move):
                        
                            score += -5
                            break  # Assuming only one block is counted per piece

        return score
        
    
    def avengibility(self, board, move):
        board.push(move)
        opponent_moves = list(board.legal_moves)
        for opp_move in opponent_moves:
            if opp_move == board.is_capture(opp_move):
                blowback = True
                board.push(opp_move)
                # Check if there's any of your moves that can capture the square where the opponent's piece moved
                revenges = [m for m in board.legal_moves if m.to_square == opp_move.to_square]
                if revenges:
                    can_avenge = True  # There exists a move that can avenge the moved piece
                board.pop()
                if can_avenge:
                    avengerscore = 10 * len(revenges)
                else:
                    avengerscore = 0
                if blowback == True:
                    avengerscore += -10
        return avengerscore
    
    
    def openlock(self, last_move):
        if last_move:
            self.punishment_squares.append(last_move.to_square)
        return punishment_squares  
    
    def quit_engine(self):
        self.engine.quit()
```
