```python
#DEFENSIVEBOT

class Defend2Win:
    def __init__(self):
        self.engine_path = "F:\\DemiDev\\Applications\\StockFish\\stockfish\\stockfish-windows-x86-64-avx2.exe"
        self.engine = chess.engine.SimpleEngine.popen_uci(self.engine_path)
        
    def piece_value(self, piece):
        """Assigns value to pieces with a higher penalty for losing non-pawns."""
        if piece is None:
            return 0
        value = {chess.PAWN: 1, chess.KNIGHT: 6, chess.BISHOP: 8, chess.ROOK: 8, chess.QUEEN: 10, chess.KING: 0} 
        return value.get(piece.piece_type, 0)
        
    def evaluate_risk(self, board):
        """Evaluates the board to prioritize risk mitigation."""
        score = 0
        for piece_type in chess.PIECE_TYPES:
            for color in [chess.WHITE, chess.BLACK]:
                piece_list = board.pieces(piece_type, color)
                for piece_square in piece_list:
                    piece = board.piece_at(piece_square)
                    if piece:
                        value = self.piece_value(piece)
                        # Adjust the score based on piece value and color
                        score += value if color == chess.WHITE else -value
        return score


    def select_move(self, board, move_history):
        """Selects the move that minimizes risk."""
        best_move = None
        best_score = float('inf')  # Start with infinity since we're minimizing risk
        capture_penalty = -50  # Penalty for losing pieces
        non_pawn_penalty = -100  # Higher penalty for losing valuable pieces
        victory_bonus = 1000  # Reward for winning
        
        for move in board.legal_moves:
            board.push(move)
            if board.is_game_over():
                if board.is_checkmate():
                    score = victory_bonus  # Winning the game is the highest priority
                else:
                    score = 0  # No significant change for draws or stalemate
            else:
                score = self.evaluate_risk(board)
                if board.is_capture(move):
                    captured_piece = board.piece_at(move.to_square)
                    score += non_pawn_penalty if captured_piece.piece_type != chess.PAWN else capture_penalty
            
            if score < best_score:  # Looking for the move with the lowest risk
                best_score = score
                best_move = move
            board.pop()

        return best_move

    def quit_engine(self):
        self.engine.quit()
```

```python
#PEOPLE'S CHAMP 1.0

class PeopleChamp:
    def __init__(self):
        self.engine_path = "F:\\DemiDev\\Applications\\StockFish\\stockfish\\stockfish-windows-x86-64-avx2.exe"
        self.engine = chess.engine.SimpleEngine.popen_uci(self.engine_path)
        self.contributors_list = ['b2d2', 'a2a4', 'd3f5', 'd2b4', 'd4d5', 'c4e5', 'e4e5', 'd5c4', 'e4e5', 'e5f3', 'f3c6', 'c6c3', 'c3f6', 'g1f3', 'g8h6', 'a2a3', 'h8g8', 'd2d4', 'e2e4', 'd7b2', 'h8g3', 'e3c7', 'f3e5', 'b5f3', 'a3a7', 'h7g2', 'a8b8', 'h8c8', 'f2a5', 'h4f6', 'a1f5', 'a2g1', 'h8h8', 'f3b8', 'd4h7', 'a1a1', 'b2f2', 'f2a2', 'h6b1', 'c2a4', 'e7a3', 'g5e6', 'c1c5', 'c4a5', 'c7h5', 'a7h8', 'a6h5', 'h6e4', 'c1g8', 'g6h3', 'd5f4', 'g5c2', 'g3b7', 'f7d2', 'h3d2', 'f5d4', 'e6h1', 'f4g2', 'h4a7', 'g4h6', 'h8d4', 'a4g5', 'f3a6', 'c2c3', 'e8e1', 'a2b8', 'b7g7', 'e2e4', 'c8f4', 'f1h3', 'c8c2', 'e8h7', 'h2h1', 'f8f5', 'c5a6', 'a4h1', 'e3b1', 'h8h1', 'd4a1', 'h1a4', 'a6h7', 'e2d8', 'd8d5', 'h2h4', 'f4d6', 'd1f6', 'b4b3', 'h4g6', 'a5d4', 'c8f6', 'c3f2', 'g7g5', 'b7f6', 'g3f7', 'h8d6', 'd3f7', 'e6e5', 'c1b7', 'b1c1', 'c6a7', 'g6e1', 'd7b6', 'e1a2', 'b7c1', 'c2b8', 'b2b1', 'c3b4', 'd1h1', 'a5f6', 'b8h6', 'e3g2', 'e2e1', 'f2a8', 'f5f1', 'c4f1', 'f3f8', 'a3f2', 'b6e1', 'b6f4', 'a7d4', 'b2d8', 'd8a4', 'c4d4', 'e8f1', 'f7d4', 'a7d7', 'b2b3', 'a6e8', 'a5g8', 'a4g6', 'b6a3', 'a6f8', 'd3f2', 'e7d6', 'f4e4', 'g6c7', 'g6a3', 'a4c3', 'g4f1', 'g7f1', 'e2h8', 'b6b3', 'c5f7', 'b3e8', 'e2d2', 'f2b3', 'c6c1', 'h5c7', 'h3b7', 'a5f8', 'g4f2', 'f1g1', 'd5c6', 'e8b3', 'h3a3', 'c6c2', 'd8b5', 'e3a7', 'g3a7', 'c1h2', 'f4b2', 'd6d2', 'b6c4', 'b6f5', 'g4f3', 'f5b4', 'e3d7', 'e5c2', 'c2e5', 'b8d7', 'a1a4', 'a2a5', 'c3b8', 'g2g4', 'h6a7', 'h1h2', 'h1a3', 'h5b6', 'g7f3', 'h7d1', 'h5h3', 'f8h3', 'h8d8', 'e6f2', 'a1d7', 'e3e2', 'h2d8', 'g4c3', 'c6d4', 'b8b2', 'g3f5', 'c8d3', 'f2e2', 'h4f3', 'f8a2', 'f3g3', 'f7b8', 'a2a6', 'h2f8', 'e8f7', 'f8b8', 'c4h6', 'a4b6', 'g8d4', 'd4a7', 'f4h8', 'g8h7', 'd8b7', 'c1a1', 'e1h6', 'e5g4', 'g2a2', 'c3c8', 'b2d3', 'a2b4', 'd1b8', 'g4g1', 'c8g1', 'd3a6', 'b3h7', 'e7g7', 'h6f2', 'a8h5', 'd8h1', 'h3h5', 'h3e1', 'd1e3', 'f3a1', 'c5a2', 'c6b6', 'g8c1', 'e4h8', 'f6e6', 'h8e2', 'g2e6', 'c7h8', 'b1a2', 'c5f8', 'g3f2', 'c1d6', 'c8b8', 'f8e7', 'd3e1', 'e6e1', 'b2b6', 'h7h2', 'c5a4', 'd7d5', 'e7f2', 'e4b7', 'c7a3', 'e1e3', 'd8h7', 'd3g8', 'f5a4', 'c1h7', 'f4c5', 'h3e8', 'a1g5', 'b6c3', 'e3f6', 'g3d8', 'h4c5', 'c7f2', 'a1c3', 'c6b5', 'e1g2', 'f5f2', 'e8f8']

    def get_top_moves_from_stockfish(self, board, top_n=10):
        top_moves = []
        try:
            with self.engine.analysis(board, multipv=top_n, limit=chess.engine.Limit(time=1)) as analysis:
                for info in analysis:
                    if "pv" in info and info["pv"]:
                        # Iterate over the first three moves of each PV, if available
                        for move in info["pv"][:3]:  # Get up to the first three moves
                            if move not in top_moves:
                                top_moves.append(move)
                            if len(top_moves) >= top_n * 3:  # Adjust the limit to account for 3 moves per PV
                                break
                    if len(top_moves) >= top_n * 3:  # Ensure we break out if we've collected enough moves
                        break
        except Exception as e:
            print(f"An error occurred during Stockfish analysis: {e}")
        return top_moves

    
    def piece_value(self, piece):
        """Assigns value to pieces with a higher penalty for losing non-pawns."""
        if piece is None:
            return 0
        value = {chess.PAWN: 1, chess.KNIGHT: 6, chess.BISHOP: 8, chess.ROOK: 8, chess.QUEEN: 10, chess.KING: 0} 
        return value.get(piece.piece_type, 0)

    def evaluate_move(self, board, move):
        score = 0
        capture_bonus = 35
        non_pawn_bonus = 100
        loss_penalty = -35
        if move not in board.legal_moves:
            return -float('inf')
            
        # Make the move on a copy of the board to evaluate its consequences
        board_copy = board.copy(stack=False)
        board_copy.push(move)
    
        # Check if the move captures an opponent's piece
        if board.is_capture(move):
            captured_piece = board_copy.piece_at(move.to_square)
            if captured_piece.piece_type != chess.PAWN: 
                score += non_pawn_bonus  # Additional bonus for capturing non-pawn pieces
            else:
                score += capture_bonus  # Regular bonus for capturing a pawn
        
        # Evaluate if the move exposes the piece to capture
        attackers_after_move = board_copy.attackers(not board.turn, move.to_square)
        if attackers_after_move:
            moved_piece = board_copy.piece_at(move.to_square)
            score += self.piece_value(moved_piece) * loss_penalty  # Apply penalty based on the value of the moved piece
        
        return score


    def select_move(self, board, move_history):
        if board.is_game_over():
            print("Game over detected.")
            return None  # Or handle according to your game's logic
        
        top_moves = self.get_top_moves_from_stockfish(board, 10)
        prioritized_moves = []
    
        # Convert moves to UCI format and prioritize based on contributors list
        for move in top_moves:
            if move.uci() in self.contributors_list:
                prioritized_moves.append(move)
    
        # Use prioritized moves if available; otherwise, continue with top_moves
        moves_to_evaluate = prioritized_moves if prioritized_moves else top_moves
            

        move_scores = []
        for move in top_moves:
            move_score = self.evaluate_move(board, move)
            move_scores.append((move, move_score))
        
        # Sort moves by their adjusted score
        move_scores.sort(key=lambda x: x[1], reverse=True)  # Sort moves based on score
        
        # Select the best move after scoring adjustments
        best_move = move_scores[0][0] if move_scores else None
        return best_move

    def quit_engine(self):
        self.engine.quit()
```

```python
#PEOPLE'S CHAMP 2.0
class PeopleChampV2: 
    def __init__(self):
        self.engine_path = "F:\\DemiDev\\Applications\\StockFish\\stockfish\\stockfish-windows-x86-64-avx2.exe"
        self.engine = chess.engine.SimpleEngine.popen_uci(self.engine_path)
        self.target_squares = {
            chess.PAWN: ['e4', 'g5', 'f3'], 
            chess.KNIGHT: ['f3', 'c4', 'a2'], 
            chess.BISHOP: ['c4', 'c6', 'a3'], 
            chess.ROOK: ['d2', 'e2', 'f2', 'd3'],
            chess.QUEEN: ['d2', 'e5', 'f2', 'c6', 'g1', 'a2'],
            chess.KING: ['d1', 'e1', 'f1', 'g1']
        }

    
    def get_top_moves_from_stockfish(self, board, top_n=10):
        top_moves = []
        try:
            with self.engine.analysis(board, multipv=top_n, limit=chess.engine.Limit(time=3)) as analysis:
                for info in analysis:
                    if "pv" in info and info["pv"]:
                        # Iterate over the first three moves of each PV, if available
                        for move in info["pv"][:3]:  # Get up to the first three moves
                            if move not in top_moves:
                                top_moves.append(move)
                            if len(top_moves) >= top_n * 3:  # Adjust the limit to account for 3 moves per PV
                                break
                    if len(top_moves) >= top_n * 3:  # Ensure we break out if we've collected enough moves
                        break
        except Exception as e:
            print(f"An error occurred during Stockfish analysis: {e}")
        return top_moves

    def piece_value(self, piece):
        """Assigns value to pieces with a higher penalty for losing non-pawns."""
        if piece is None:
            return 0
        value = {chess.PAWN: 1, chess.KNIGHT: 6, chess.BISHOP: 8, chess.ROOK: 8, chess.QUEEN: 10, chess.KING: 0} 
        return value.get(piece.piece_type, 0)


    def evaluate_move(self, board, move):
        score = 0
        capture_bonus = 35
        non_pawn_bonus = 80
        loss_penalty = -35
        if move not in board.legal_moves:
            return -float('inf')
            
        board_copy = board.copy(stack=False)
        board_copy.push(move)
    
        if board.is_capture(move):
            captured_piece = board_copy.piece_at(move.to_square)
            if captured_piece.piece_type != chess.PAWN: 
                score += non_pawn_bonus
            else:
                score += capture_bonus
        
        attackers_after_move = board_copy.attackers(not board.turn, move.to_square)
        if attackers_after_move:
            moved_piece = board_copy.piece_at(move.to_square)
            score += self.piece_value(moved_piece) * loss_penalty

        for piece_type, target_squares in self.target_squares.items():
            for target_square_str in target_squares:
                square_index = chess.parse_square(target_square_str)
                piece_at_target = board_copy.piece_at(square_index)
                if piece_at_target and piece_at_target.piece_type == piece_type:
                    score += 100

        return score


    def select_move(self, board, move_history):
        if board.is_game_over():
            print("Game over detected.")
            return None  # Or handle according to your game's logic
        
        top_moves = self.get_top_moves_from_stockfish(board, 10)
        prioritized_moves = []
    
        # Use prioritized moves if available; otherwise, continue with top_moves
        moves_to_evaluate = prioritized_moves if prioritized_moves else top_moves
            

        move_scores = []
        for move in top_moves:
            move_score = self.evaluate_move(board, move)
            move_scores.append((move, move_score))
        
        # Sort moves by their adjusted score
        move_scores.sort(key=lambda x: x[1], reverse=True)  # Sort moves based on score
        
        # Select the best move after scoring adjustments
        best_move = move_scores[0][0] if move_scores else None
        return best_move

    def quit_engine(self):
        self.engine.quit()
            
```

```python
#KING'S ASSASSIN
class KingAssassin:
    def __init__(self):
        self.engine_path = "F:\\DemiDev\\Applications\\StockFish\\stockfish\\stockfish-windows-x86-64-avx2.exe"
        self.engine = chess.engine.SimpleEngine.popen_uci(self.engine_path)
        self.depth_check_bonus = {1: 1000, 2: 250, 3: 150, 4: 100, 5: 35}
        self.time_limit = 0.2
        
    def piece_value(self, piece):
        """Assigns value to pieces with a higher penalty for losing non-pawns."""
        if piece is None:
            return 0
        value = {chess.PAWN: 1, chess.KNIGHT: 3, chess.BISHOP: 6, chess.ROOK: 6, chess.QUEEN: 10, chess.KING: 0} 
        return value.get(piece.piece_type, 0)

    def evaluate_move(self, board, move):
        self.start_time = time.time()
        score = 0
        # Make the move on a copy of the board to evaluate its consequences
        board_copy = board.copy(stack=False)
        board_copy.push(move)

        # Reward for putting the opponent's King in check
        if board_copy.is_check():
            score += self.depth_check_bonus[1]

        # Reward for taking higher level pieces and penalize for losing non-pawn pieces
        if board.is_capture(move):
            captured_piece = board_copy.piece_at(move.to_square)
            capturing_piece = board.piece_at(move.from_square)
            score += self.piece_value(captured_piece)

            # Check if the capturing move exposes one of our valuable pieces
            if capturing_piece.piece_type != chess.PAWN:
                score -= self.piece_value(capturing_piece) * 0.699  # Penalize but less than the value of the piece itself
                
        # Evaluate win/loss condition
        if board_copy.is_game_over():
            if board_copy.result() == "1-0" if board.turn == chess.WHITE else "0-1":
                score += 1000  # Greatly reward winning
            else:
                score -= 1000  # Penalize losing or drawing

        
        score += self.assesstheking(board_copy)
        score += self.evaluate_future_checks(board_copy, 1)
        
        return score
        
    def assesstheking(self, board):
        # Count the King's bodyguards
        king_square = board.king(board.turn)
        bodyguards = 0
        safe_position = True  # Assume the king is in a safe position initially

        # Correct calculation for king's rank and file
        king_rank = chess.square_rank(king_square)
        king_file = chess.square_file(king_square)
                 
        for i in range(-2, 3):  # From -3 to 3 including
            for j in range(-2, 3):  # From -3 to 3 including
                if i == 0 and j == 0:
                    continue  # Skip the king's square itself
                neighbor_rank = king_rank + i
                neighbor_file = king_file + j
                
                # Ensure the square is on the board
                if 0 <= neighbor_rank < 8 and 0 <= neighbor_file < 8:
                    neighbor_square = chess.square(neighbor_file, neighbor_rank)
                    piece = board.piece_at(neighbor_square)
                    # Count as a bodyguard if the piece is friendly and not the king
                    if piece and piece.piece_type != chess.KING and piece.color == board.turn:
                        bodyguards += 1
        if board.is_attacked_by(not board.turn, king_square):
            safe_position = False
    
        score_adjustment = 0
        if bodyguards < 5:
            score_adjustment += (5 - bodyguards) * 500
        if not safe_position:
            score_adjustment -= 1000
        return score_adjustment



    def evaluate_future_checks(self, board, current_depth):
        if time.time() - self.start_time > self.time_limit:
            return 0
            
        if current_depth >= 5:  # Base case: stop recursion at depth 5
            return 0
    
        score = 0
        for future_move in board.legal_moves:
            future_board = board.copy(stack=False)
            future_board.push(future_move)
    
            if future_board.is_check():
                score += self.depth_check_bonus[current_depth + 1]
    
            # Recursive call to evaluate further checks at the next depth
            score += self.evaluate_future_checks(future_board, current_depth + 1)
    
        return score

    def evaluate_endgame_move(self, board, move):
        score = 0
        if board.piece_at(move.to_square) and board.piece_at(move.to_square).piece_type == chess.PAWN and (chess.square_rank(move.to_square) in [0, 7]):
            score += 200  # High score for promoting a pawn
        if board.gives_check(move):
            score += 600  # Bonus for moves that give check
        return score
       
    
        # Count the King's bodyguards
        king_square = board.king(board.turn)
        bodyguards = 0
        for i in range(-2, 3):  # From -3 to 3 including
            for j in range(-2, 3):  # From -3 to 3 including
                if i == 0 and j == 0:
                    continue  # Skip the king's square itself
                neighbor_rank = king_rank + i
                neighbor_file = king_file + j
                
                # Ensure the square is on the board
                if 0 <= neighbor_rank < 8 and 0 <= neighbor_file < 8:
                    neighbor_square = chess.square(neighbor_file, neighbor_rank)
                    piece = board.piece_at(neighbor_square)
                    # Count as a bodyguard if the piece is friendly and not the king
                    if piece and piece.piece_type != chess.KING and piece.color == board.turn:
                        bodyguards += 1
        
        # If the King has less than optimal bodyguards, adjust the score to prioritize moving pieces back to protect him
        if bodyguards < 5:
            score += (bodyguards - 2) * 500  # Example adjustment: more points for moves bringing protection
        return score

    def is_endgame(self, board):
        # Initialize piece counts for both colors
        white_pieces_count = 0
        black_pieces_count = 0
    
        # Count the pieces for each color
        for piece_type in chess.PIECE_TYPES:
            white_pieces_count += len(board.pieces(piece_type, chess.WHITE))
            black_pieces_count += len(board.pieces(piece_type, chess.BLACK))
    
        # Calculate the difference in piece count
        piece_count_difference = abs(white_pieces_count - black_pieces_count)
    
        # Define endgame as when one color has an advantage of more than 6 pieces
        return piece_count_difference > 6

    def select_move(self, board, move_history):
        legal_moves = list(board.legal_moves)
        best_move = None
        best_score = float('-inf')
        endgame = self.is_endgame(board)

        for move in legal_moves:
            if not endgame:
                score = self.evaluate_move(board, move)
                if score > best_score:
                    best_score = score
                    best_move = move
            else:
                score = self.evaluate_endgame_move(board, move)
                if score > best_score:
                    best_score = score
                    best_move = move

        return best_move

    

    def quit_engine(self):
        self.engine.quit()
```

```python
#WILD CARD

class WilDCArD:
    def __init__(self):
        self.engine_path = "F:\\DemiDev\\Applications\\StockFish\\stockfish\\stockfish-windows-x86-64-avx2.exe"
        self.engine = chess.engine.SimpleEngine.popen_uci(self.engine_path)
        self.max_score = 50
        self.depth_check_bonus = {1: 1000, 2: 250, 3: 150, 4: 100, 5: 35}
        self.time_limit = 0.2
        
    def piece_value(self, piece):
        if piece is None:
            return 0
        value = {chess.PAWN: 1, chess.KNIGHT: 3, chess.BISHOP: 3, chess.ROOK: 5, chess.QUEEN: 9, chess.KING: 0}  # Adjusted values for standardization
        return value.get(piece.piece_type, 0)
        
    def aggro_targets(self, board):
        score = 0
        current_turn = board.turn  
        threat_bonus = 150
        my_major_pieces = [chess.QUEEN, chess.ROOK, chess.BISHOP, chess.KNIGHT, chess.KING]
        my_pieces_squares = []

        for piece_type in my_major_pieces:
            my_pieces_squares.extend(board.pieces(piece_type, current_turn))

        
        threatening_pieces = set()
        for my_piece_square in my_pieces_squares:
            # Find opponent pieces that threaten your major piece
            attackers = board.attackers(not current_turn, my_piece_square)
            for attacker_square in attackers:
                if board.piece_at(attacker_square).color != current_turn:
                    threatening_pieces.add(attacker_square)
    
            
        for piece_type in chess.PIECE_TYPES:
            for color in [chess.WHITE, chess.BLACK]:
                piece_list = board.pieces(piece_type, color)
                for piece_square in piece_list:
                    piece = board.piece_at(piece_square)
                    if piece:
                        value = self.piece_value(piece)
                        # Score positively for the current player's pieces and negatively for the opponent's
                        if color == current_turn:
                            score += value
                        else:
                            score -= value
                            if piece_square in threatening_pieces:
                                score += threat_bonus
        return score

    def evaluate_future_checks(self, board, current_depth):
        if time.time() - self.start_time > self.time_limit:
            return 0
            
        if current_depth >= 5:  # Base case: stop recursion at depth 5
            return 0
    
        score = 0
        for future_move in board.legal_moves:
            future_board = board.copy(stack=False)
            future_board.push(future_move)
    
            if future_board.is_check():
                score += self.depth_check_bonus[current_depth + 1]
    
            # Recursive call to evaluate further checks at the next depth
            score += self.evaluate_future_checks(future_board, current_depth + 1)
    
        return score

    def evaluate_move(self, board, move):
        self.start_time = time.time()
        score = 0
        board.push(move)

        # Reward for win odds improvement and capturing non-pawn pieces
        if board.is_capture(move):
            captured_piece = board.piece_at(move.to_square)
            if captured_piece.piece_type != chess.PAWN:
                score += 10  # Assign points for capturing a non-pawn

        # Check for winning condition
        if board.is_checkmate():
            score += 50  # Winning move gets the highest score

        board.pop()
        return score
        
    def assesstheking(self, board):
        # Count the King's bodyguards
        king_square = board.king(board.turn)
        bodyguards = 0
        safe_position = True  # Assume the king is in a safe position initially

        # Correct calculation for king's rank and file
        king_rank = chess.square_rank(king_square)
        king_file = chess.square_file(king_square)
                 
        for i in range(-2, 3):  # From -3 to 3 including
            for j in range(-2, 3):  # From -3 to 3 including
                if i == 0 and j == 0:
                    continue  # Skip the king's square itself
                neighbor_rank = king_rank + i
                neighbor_file = king_file + j
                
                # Ensure the square is on the board
                if 0 <= neighbor_rank < 8 and 0 <= neighbor_file < 8:
                    neighbor_square = chess.square(neighbor_file, neighbor_rank)
                    piece = board.piece_at(neighbor_square)
                    # Count as a bodyguard if the piece is friendly and not the king
                    if piece and piece.piece_type != chess.KING and piece.color == board.turn:
                        bodyguards += 1
        if board.is_attacked_by(not board.turn, king_square):
            safe_position = False
    
        score_adjustment = 0
        if bodyguards < 5:
            score_adjustment += (5 - bodyguards) * 800
        if not safe_position:
            score_adjustment -= 2000
        return score_adjustment

    

    def select_move(self, board, move_history):
        best_move = None
        best_score = -1

        for move in board.legal_moves:
            score = self.evaluate_move(board, move)
            score += self.aggro_targets(board)
            score += self.assesstheking(board)
            score += self.evaluate_future_checks(board, 1)
            
            # Iteratively check for a better score
            if score > best_score:
                best_score = score
                best_move = move

                # Exit early if the maximum score is achieved
                if best_score >= self.max_score:
                    break

        return best_move

    def quit_engine(self):
        self.engine.quit()
```
